* 312 locationType :replace Shared Call Appearance with Secondary Device
## **Remove unwanted email ids**
* 312 location Type: Remove unwanted email ids (replace with nan)
## **Treating Nan of column 147 relatedCallIdReason with 267 vpDialingfacResult 312 locationType**

* Replace nan of 147 as per the below condition:
* Replace nan by concatenating 312 and 267 data provided both are not null
* Replace nan with 312 data if 267 is nan or  with 267 data if 312 is nan
